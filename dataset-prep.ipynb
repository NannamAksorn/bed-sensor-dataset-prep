```python
from glob import glob
import pandas as pd
import os
```

```python
rooms = glob("/Volumes/MyPassport/Old HDD/convertfile/*")
rooms.remove('/Volumes/MyPassport/Old HDD/convertfile/402_1')
rooms
```

```python
for roompath in rooms:
    room = os.path.basename(roompath)
    sendats = glob(roompath + "/S*txt.xlsx")
    csv_paths = glob("/Volumes/MyPassport/Old HDD/edited_key/{}/S*.csv".format(room))
        
    for sendat in sendats:
        try:
            df = pd.read_excel(sendat, header=None)
            df.columns = ["p0", "w0", "p1", "w1"]
            df['class'] = 9
        except Exception:
            continue
        
        found = False
        for csv_path in csv_paths:
            if os.path.basename(csv_path)[:33] == os.path.basename(sendat)[:33]:
                print('found')
                found = True
                break
        if not found:
            print('not found', os.path.basename(sendat)[:33])
            continue
            
        try:
            df_key = pd.read_csv(csv_path, header=None)
            df_key = df_key.values
        except Exception:
            continue
            
        for i in range(len(df_key) -1 ):
            index_start, key = df_key[i]
            index_end, _ = df_key[i+1]
            index_start = int(index_start)
            index_end = int(index_end)
            if int(key) == 0:
                key = 9
            df['class'][index_start:index_end] = key
            
        date = os.path.basename(sendat)[18:33]
        if int(date[:4]) > 2500:
            date = str(int(date[:4]) - 543) + date[4:]

        output_path = "./output/{}/Bed_{}_{}.csv".format(room,date,room)
        if not os.path.exists(os.path.dirname(output_path)):
            os.mkdir(os.path.dirname(output_path))
        df.to_csv(output_path, sep=',', encoding='utf-8',index=False)
        print(sendat)
```
